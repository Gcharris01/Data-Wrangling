```python
# need missing library
import matplotlib.pyplot as plt
import numpy as np # Importing the numpy library and aliasing it as np

# Dataset : Feature (sq. ft) and target (price)
X = np.array([1500, 1600, 1700, 1900]).reshape(-1, 1)
y = np.array([300000, 320000, 340000, 360000]) # Fixed: Removed the extra element

# Intercept
X_b = np.c_[np.ones((X.shape[0], 1)), X]

# Cost function, add meaningful names to the variables
def compute_cost(X, y, theta):
    i = len(y)
    ii = X.dot(theta)
    iii = (1 / (2 * i)) * np.sum((ii -y) ** 2)
    return iii

# Add an two additional sets, 0<sub>0 and 0<sub>1
theta_values = [
    np.array([100000, 150]),
    np.array([120000, 130]),
    np.array([110000, 140])
]

# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color="blue", label="Data Points") # Changed label to "Data Points"

# Create all new values for the variable passed into the function
for i, theta in enumerate(theta_values):

    # Compute the cost of the parameters
    cost = compute_cost(X_b, y, theta)

    # Print only the cost
    print(f"The cost for theta {i+1}: 0<sub>0 = {theta[0]}, 0<sub>1> = {theta[1]}) is: {cost}")

    # Compute predictions
    hx = X_b.dot(theta)

    # Plot the fitted straight line
    plt.plot(X, hx, label=f"Fitted Line {i+1}: (θ0 = {theta[0]}, θ1 = {theta[1]})") # Changed label format


# Add Description to labels and title to reflect the values and purpose
plt.xlabel("Area (sq. ft)") # Changed x-axis label
plt.ylabel("Price") # Changed y-axis label
plt.title("House Price Prediction") # Changed title
plt.legend()
plt.grid(True)
plt.show()
```
